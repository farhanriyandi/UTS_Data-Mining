```python
import numpy as np 
import pandas as pd  
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_excel('format.xlsx')
```

```python
data
```

```python
data['Transaksi'] = data['Transaksi'].astype('str')
```

```python
data.dropna(axis=0, subset=['Transaksi'], inplace=True) 
```

```python
data = data.set_index('Transaksi')
```

```python
data
```

```python
def hot_encode(x):
    if(x<=0): 
        return 0 
    if(x>=1): 
        return 1
```

```python
data = data.applymap(hot_encode)
data
```

```python
frq_items = apriori(data, min_support=0.5, use_colnames=True) 
rules = association_rules(frq_items, metric = "lift", min_threshold=1) 
rules = rules.sort_values(['confidence', 'lift'], ascending=[False, False]) 
#print(rules.head())
rules.head()
```
